```python
def is_balanced(expression):
    stack = []
    pairs = {')': '(', '}': '{', ']': '['}

    for char in expression:
        # If opening brace, push to stack
        if char in "({[":
            stack.append(char)

        # If closing brace
        elif char in ")}]":
            if not stack:
                return False
            top = stack.pop()
            if pairs[char] != top:
                return False

    # If stack is empty → balanced
    return len(stack) == 0


# Example usage
exp = input("Enter expression: ")

if is_balanced(exp):
    print("Braces are balanced ✅")
else:
    print("Braces are NOT balanced ❌")
```
