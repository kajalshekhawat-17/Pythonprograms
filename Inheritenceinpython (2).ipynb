```python
# Parent class
class Amphibian:
    def __init__(self, name):
        self.name = name

    def habitat(self):
        print(self.name, "lives both on land and in water")

    def movement(self):
        print(self.name, "can move")


# Child class 1
class Frog(Amphibian):
    def movement(self):
        print(self.name, "jumps and swims")


# Child class 2
class Salamander(Amphibian):
    def movement(self):
        print(self.name, "walks and swims")


# Creating objects
a1 = Frog("Frog")
a2 = Salamander("Salamander")

# Calling methods
a1.habitat()
a1.movement()

a2.habitat()
a2.movement()
```
