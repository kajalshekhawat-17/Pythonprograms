#**List and conditions (Drinks)**

```python
drinks = ["tea", "coffee", "juice", "milk"]

name = input("Enter drink name: ")

if name in drinks:
    print("Drink is available")
else:
    print("Drink is not available")
```

#**Average of list**

```python
n = int(input("Enter number of elements: "))

numbers = []

for i in range(n):
    num = int(input("Enter number: "))
    numbers.append(num)

total = 0
for num in numbers:
    total = total + num

average = total / n

print("Average =", average)
```

#**Average of five students**

```python
# take marks of 5 students
m1 = int(input("Enter marks of student 1: "))
m2 = int(input("Enter marks of student 2: "))
m3 = int(input("Enter marks of student 3: "))
m4 = int(input("Enter marks of student 4: "))
m5 = int(input("Enter marks of student 5: "))

# calculate total
total = m1 + m2 + m3 + m4 + m5

# calculate average
average = total / 5

print("Average marks =", average)
```
