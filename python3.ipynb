#**For loops in python**

```python
name = "Kajal"

for letter in name:
    print(letter)
```

#**Same message multiple times**

```python
message = "Hello"
i = 1

while i <= 5:
    print(message)
    i += 1
```

#**While loop with message**

```python
msg = input("Enter message: ")
n = int(input("Enter number of times: "))

i = 1
while i <= n:
    print(msg)
    i = i + 1
```

#**Print letters of a word**

```python
word = input("Enter a word: ")

i = 0
while i < len(word):
    print(word[i])
    i = i + 1
```
