```python
def isPal(x):
    """Returns True if the list is a palindrome"""
    temp = x.copy()      # make a copy
    temp.reverse()       # reverse the copy
    return temp == x     # compare


def silly(n):
    """Takes n inputs and checks if they form a palindrome"""
    result = []   # create list once

    for i in range(n):
        elem = input("Enter element: ")
        result.append(elem)

    if isPal(result):
        print("Yes")
    else:
        print("No")


# function call
silly(3)
```
