```python
import dill

class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        return f"{self.name} scored {self.marks}"

# Create object
s1 = Student("Sameera", 85)

# Save object
with open("student.pkl", "wb") as f:
    dill.dump(s1, f)

# Delete object
del s1

# Load object
with open("student.pkl", "rb") as f:
    s1 = dill.load(f)

print(s1.display())
```
